```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

# Carrega a tabela imortação
exp_soja=pd.read_csv('/home/andre301267/git/Pricing-Fertilizer/DB/temp/exp_soja_ncm_cat.csv').reset_index(drop=True)
exp_soja.columns
```

### Identifica as colunas quantitativas

```python
col_quanti=['kg','US$_fob']
```

### Avalia impacto e elimina registros com quantidade igual a zero

```python
# Avalia impacto na eliminação de registros com quantidade zero
for q in col_quanti:
    print(f"{q} possui {round(100*exp_soja[exp_soja[q]==0].shape[0]/exp_soja.shape[0],1)} % dos registros com valor igual a zero.")
```

```python
for q in col_quanti:
    par=exp_soja[exp_soja[q]==0][[x for x in col_quanti if x not in [q]]].sum()
    tot=exp_soja[exp_soja[q]!=0][[x for x in col_quanti if x not in [q]]].sum()
    print('Qdo ',q,'= 0:')
    print(f"{100*(par/tot)} % do total.")
    print()
```

```python
# Descarta registros com valor igual a zero
for c in col_quanti:
    exp_soja=exp_soja[exp_soja[c]>0]
```

```python
exp_soja=exp_soja.groupby(by=['year','month','UF'])[['kg','US$_fob']].sum().reset_index()
```

```python
exp_soja=exp_soja.assign(Product='Soy Group')
```

```python
col_quanti=['kg','US$_fob']
```

```python
for c in col_quanti:
    sns.displot(kind='kde', data=exp_soja, x=c, col='Product', height=2, aspect=2,\
                log_scale=True, facet_kws=dict(sharey=False)).fig.suptitle('Brazilian Soy Group Export', y=1.05)

#plt.savefig('/home/andre301267/git/Pricing-Fertilizer/DB/Comexstat/graphs/fertilizers_imports.png')
plt.show()
```

```python
for c in col_quanti:
    exp_soja.insert(len(exp_soja.columns),'log_'+c, np.log10(exp_soja[c]))
exp_soja
```

```python
from sklearn.cluster import KMeans as KM
KM=KM(n_clusters=2)

exp_soja=exp_soja.assign(km=KM.fit_predict(exp_soja[['log_kg','log_US$_fob']]))

for c in col_quanti:
    sns.displot(kind='kde', data=exp_soja, x=c, col='Product', height=2, aspect=2,hue='km',\
                log_scale=True, facet_kws=dict(sharey=False)).fig.suptitle('Brazilian Soy Group Export', y=1.05)

#plt.savefig('/home/andre301267/git/Pricing-Fertilizer/DB/Comexstat/graphs/fertilizers_imports.png')
plt.show()
```

```python
km_filtro=exp_soja.groupby('km').kg.mean().sort_values().reset_index().tail(1).km.values[0]
km_filtro
```

```python
exp_soja=exp_soja[exp_soja.km==km_filtro]
```

```python
exp_soja=exp_soja[[ 'year','month', 'Product', 'UF', 'kg', 'US$_fob', 'km']]
```

```python
exp_soja.to_csv('/home/andre301267/git/Pricing-Fertilizer/DB/temp/exp_soja_ncm_quanti.csv',index=False)
```
