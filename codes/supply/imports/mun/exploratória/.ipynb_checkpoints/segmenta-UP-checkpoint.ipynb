```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

imp_fert=pd.read_csv('~/git/BD/temp/imp_fert_ncm_quanti.csv')

up_est=imp_fert[(imp_fert.km==imp_fert.km.min())&(imp_fert.nick=='NP')]

up_est_o=up_est[(up_est.CO_ANO!=2024)]

up_est_o_ano_pais=up_est_o.groupby(by=['CO_ANO','NO_PAIS']).KG_LIQUIDO.sum().reset_index().sort_values('KG_LIQUIDO')

l=up_est_o_ano_pais.groupby('NO_PAIS').KG_LIQUIDO.sum().reset_index().sort_values('KG_LIQUIDO').tail(5).NO_PAIS.values

up_est_o_ano_pais.loc[~up_est_o_ano_pais.NO_PAIS.isin(l),'NO_PAIS']='outros'

title='Estimativa de Importao de UP no Brasil'

up_est_o_ano_pais_ano_pais_kg=up_est_o_ano_pais.groupby(by=['CO_ANO','NO_PAIS']).KG_LIQUIDO.sum().reset_index()

sns.relplot(data=up_est_o_ano_pais_ano_pais_kg,x='CO_ANO',y='KG_LIQUIDO',hue='NO_PAIS',height=2,aspect=4).fig.suptitle(title,y=1.1)
plt.show()
```

```python
up_est_o_ano_pais_ano=up_est_o_ano_pais.groupby(by=['CO_ANO']).KG_LIQUIDO.sum().reset_index()

sns.lmplot(data=up_est_o_ano_pais_ano,x='CO_ANO',y='KG_LIQUIDO',height=2,aspect=4).fig.suptitle(title,y=1.1)
plt.show()
```

```python
up_est_o_ano_pais_ano_china=up_est_o_ano_pais[(up_est_o_ano_pais.NO_PAIS=='China')].groupby(by=['CO_ANO','NO_PAIS']).KG_LIQUIDO.sum().reset_index()

sns.lmplot(data=up_est_o_ano_pais_ano_china,x='CO_ANO',y='KG_LIQUIDO',hue='NO_PAIS',height=2,aspect=4,palette='Set1').fig.suptitle(title,y=1.1)
plt.show()
```

```python
up_est_o_ano_pais_preo_cif=up_est_o.groupby(by=['CO_ANO','NO_PAIS']).preo_cif.mean().reset_index()



l=up_est_o_ano_pais_preo_cif.groupby('NO_PAIS').preo_cif.sum().reset_index().sort_values('preo_cif').tail(5).NO_PAIS.values

up_est_o_ano_pais_preo_cif.loc[~up_est_o_ano_pais_preo_cif.NO_PAIS.isin(l),'NO_PAIS']='outros'

sns.relplot(data=up_est_o_ano_pais_preo_cif,x='CO_ANO',y='preo_cif',hue='NO_PAIS',height=2,aspect=4).fig.suptitle(title,y=1.1)
plt.show()
```

```python
up_est_o_ano_preo_cif=up_est_o_ano_pais_preo_cif.groupby(by=['CO_ANO']).preo_cif.mean().reset_index()

sns.lmplot(data=up_est_o_ano_preo_cif,x='CO_ANO',y='preo_cif',height=2,aspect=4).fig.suptitle(title,y=1.1)
plt.show()
```

```python
up_est_o_ano_preo_cif_china=up_est_o_ano_pais_preo_cif[(up_est_o_ano_pais_preo_cif.NO_PAIS=='China')].groupby(by=['CO_ANO','NO_PAIS']).preo_cif.mean().reset_index()

sns.lmplot(data=up_est_o_ano_preo_cif_china,x='CO_ANO',y='preo_cif',hue='NO_PAIS',height=2,aspect=4,palette='Set1').fig.suptitle(title,y=1.1)
plt.show()
```

```python
up_est_o_china_preo_frete=up_est_o[up_est_o.NO_PAIS=='China'].groupby(by=['CO_ANO','NO_PAIS']).preo_frete.mean().reset_index()

sns.lmplot(data=up_est_o_china_preo_frete,x='CO_ANO',y='preo_frete',height=2,aspect=4,hue='NO_PAIS',palette='Set1').fig.suptitle(title,y=1.1)
plt.show()
```

```python
up_est_o_mes=up_est_o.groupby(by=['CO_MES']).KG_LIQUIDO.count().reset_index().rename(columns={'KG_LIQUIDO':'#desembarques'})
```

```python
sns.relplot(data=up_est_o_mes.rename(columns={'CO_MES':'MS'}),x='MS',y='#desembarques',height=2,aspect=4).fig.suptitle(title,y=1.1)
plt.show()
```
