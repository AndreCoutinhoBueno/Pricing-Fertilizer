```python
a=q.drop(columns='code')
b=q.code.str.split('-',expand=True).rename(columns={0:'nick',1:'SG_UF_NCM'})
q=pd.concat([a,b],axis=1)
q
```

```python
qg=q.groupby(by=['CO_ANO','CO_MES','nick','SG_UF_NCM'])[['oc','oc_pred','KG_LIQUIDO','kg_pred']].sum().reset_index()
qg
```

```python
qg.loc[qg.oc>1,'oc']=1
qg.loc[qg.oc_pred>1,'oc_pred']=1
```

```python
metrics.accuracy_score(qg.oc,qg.oc_pred)
```

```python
metrics.precision_score(qg.oc,qg.oc_pred)
```

```python
metrics.recall_score(qg.oc,qg.oc_pred)
```

```python
qg_r=qg[(qg.oc==1)&(qg.oc_pred==1)]
qg_r
```

```python
metrics.r2_score(qg_r.KG_LIQUIDO,qg_r.kg_pred)
```
