```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as  plt
import seaborn as sns
imp_fert=pd.read_csv('~/git/BD/temp/imp_fert_quanti.csv')#.sort_values(by=['CO_ANO','CO_MES']).reset_index(drop=True)
imp_fert
```

```python
imp_fert=imp_fert[imp_fert.CO_ANO!=2024]
```

```python
imp_fert_g=imp_fert.groupby(by=['CO_MES','nick']).log_kg.sum().reset_index()

from sklearn.preprocessing import MinMaxScaler as MMS
MMS=MMS()
q=pd.DataFrame()
for nick in imp_fert_g.nick.unique():
    imp_fert_g_t=imp_fert_g[imp_fert_g.nick==nick]
    imp_fert_g_t=imp_fert_g_t.assign(log_kg_s=MMS.fit_transform(imp_fert_g_t[['log_kg']]))
    q=pd.concat([q,imp_fert_g_t])
imp_fert_g=q.copy()
imp_fert_g
```

```python
imp_fert_g=imp_fert_g[~imp_fert_g.nick.isin(['NPK_k','NP_k'])]
```

```python
sns.relplot(data=imp_fert_g,kind='line',x='CO_MES',y='log_kg_s',hue='nick')
```

```python
filtro_sub={'MAP_k':[2,7],'MOP_k':[2,7],'SAM_k':[5,10],'UREIA_k':[6,11]}
```

```python
q=pd.DataFrame()
for nick in imp_fert_g.nick.unique():
    imp_fert_g_t=imp_fert_g[imp_fert_g.nick==nick]
    imp_fert_g_t=imp_fert_g_t[(imp_fert_g_t.CO_MES>=filtro_sub[nick][0])&(imp_fert_g_t.CO_MES<=filtro_sub[nick][1])]
    q=pd.concat([q,imp_fert_g_t])
imp_fert_g_sub=q.copy()
imp_fert_g_sub
```

```python
sns.relplot(data=imp_fert_g_sub,kind='line',x='CO_MES',y='log_kg_s',hue='nick')
```

```python
imp_fert_g.loc[imp_fert_g.nick.isin(['MAP_k','MOP_k']),'CO_MES']=imp_fert_g[imp_fert_g.nick.isin(['MAP_k','MOP_k'])].CO_MES+4
imp_fert_g.loc[:,'nick']=imp_fert_g.nick.replace({'MAP_k':'MAP_k_ad4','MOP_k':'MOP_k_ad4'})
```

```python
imp_fert_g.loc[imp_fert_g.nick.isin(['SAM_k']),'CO_MES']=imp_fert_g[imp_fert_g.nick.isin(['SAM_k'])].CO_MES+1
imp_fert_g.loc[:,'nick']=imp_fert_g.nick.replace({'SAM_k':'SAM_k_ad1'})
```

```python
for ad in range(1,7):
    imp_fert_g.loc[imp_fert_g.CO_MES==12+ad,'CO_MES']=ad
```

```python
sns.relplot(data=imp_fert_g,kind='line',x='CO_MES',y='log_kg_s',hue='nick')
```
