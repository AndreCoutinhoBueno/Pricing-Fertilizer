```python
def sparse_matrix(imp_fert):
    imp_fert=imp_fert[['CO_ANO', 'CO_MES', 'CO_NCM', 'SG_UF_NCM', 'KG_LIQUIDO', 'VL_FOB']]
    
    imp_fert.insert(0,'date',pd.to_datetime(imp_fert.CO_ANO.astype(str)+'-'+imp_fert.CO_MES.astype(str)+'-1'))
    
    data_ini=imp_fert.date.min()
    data_fim=imp_fert.date.max()
    
    date_cols=['CO_ANO','CO_MES']
    cat_cols=['CO_NCM','SG_UF_NCM']
    quanti_cols=['KG_LIQUIDO','VL_FOB']
    
    # Matriz Esparsa
    imp_fert.loc[:,date_cols+cat_cols]=imp_fert[date_cols+cat_cols].astype('category')
    imp_fert=imp_fert.groupby(by=date_cols+cat_cols,observed=False)[quanti_cols].sum().reset_index()
    
    
    
    imp_fert[date_cols]=imp_fert[date_cols].astype(int)
    imp_fert[cat_cols]=imp_fert[cat_cols].astype(str)
    
    imp_fert.insert(0,'date',pd.to_datetime(imp_fert.CO_ANO.astype(str)+'-'+imp_fert.CO_MES.astype(str)+'-1'))
    
    imp_fert=imp_fert[(imp_fert.date>=data_ini)&(imp_fert.date<=data_fim)]
    
    return imp_fert
```
