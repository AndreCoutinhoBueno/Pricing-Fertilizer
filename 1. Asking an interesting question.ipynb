### Chuvas interferem na precificação dos fertilizantes?  

Precificação é a confrontação entre oferta e demanda.  

Um fator que afete de forma assimétrica a oferta ou a demanda, interfere na precificação.

A incidência de chuva certamente afeta a demanda, podendo, por exemplo, ocasionar suspensão imediatamente de operações de plantio e adubação.

As chuvas podem afetar também a oferta porém, não imediatamente, já que a oferta de fertilizantes no Brasil é majoritariamente realizada através de importações oriundas de outros continentes, via transporte marítimo. Uma vez a caminho ou mesmo prestes a embarcar, dificilmente se consegue alterar o volume pedido.

***Então as chuvas podem afetar assimetricamente a oferta e a demanda e, portanto, a precificação.***

### Demanda:  

As entregas ao consumidor das empresas importadoras de fertilizantes e registrada e enviada mensalmente para a [Associação Nacional para Difusão de Adubos - ANDA](https://www.anda.org.br), que as compilam e divulgam.  

Porém estes dados não são abertos ao público.

### Oferta:  

Em grande parte representada pelas **importações**.  

Mensalmente as quantidades desembarcadas são registradas e abertas ao público pela Receita Federal, através do site [Comexstat](http://comexstat.mdic.gov.br/).

### Objetivo do Projeto: Quantificar a influência das chuvas nas importações de fertilizantes

Num mercado competitivo, muitas vezes os preços são definidos não somente sobre os valores realizados, mas também sobre a expectativa futura da oferta e da demanda.  

Como oferta e importação são por vezes tratadas como sinônimo, a cada mês, a direção das empresas do ramo aguardam a divulgação das quantidades de fertilizantes desembarcados para poderem rever ou confirmar os preços que estão praticando.

Esse ajuste de preço é feito através da comparação entre a expetativa ou **previsão** de importação contra a real quantidade divulgada.

> Expectativa = Previsão

> Real Importado > Expectativa de Importação &rarr; Ocorrencia de Sobre-Oferta &rarr; Vetor de baixa na precificação
> 
> Vice-versa

Um novo fator que leve precisão ao cálculo da expectativa, leva precisão para a precificação.

> Precisão na Previsão &rarr; Precisão na Precificação
