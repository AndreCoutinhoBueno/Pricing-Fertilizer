```python
def corr_kg_fob_via_preço(imp_fert):

    col=imp_fert.columns

    col_quanti=['KG_LIQUIDO','VL_FOB']
    
    # Preço FOB e do Frete
    for q in col_quanti:
        imp_fert.insert(len(imp_fert.columns),'log_'+q,np.log10(imp_fert[q]))
    col_quanti
    
    ### z_score
    
    col_log=imp_fert.columns[-2:]
    col_log
    
    def z_score(imp_fert,l):
        mediana=imp_fert[l].median()
        k=len(imp_fert.columns)
        imp_fert.insert(k,'median_'+l,mediana)
        imp_fert.insert(k+1,'qd_dif'+l,(imp_fert[l]-imp_fert['median_'+l])**2)
        imp_fert.insert(k+2,'std_'+l, (imp_fert['qd_dif'+l].sum()/len(imp_fert))**(1/2))
        imp_fert.insert(k+3,'z_score_'+l, (imp_fert[l]-imp_fert['median_'+l])/imp_fert['std_'+l])
        
        return imp_fert
    
    for l in col_log:
        imp_fert=z_score(imp_fert,l)
        
    
    k=len(imp_fert.columns)
    imp_fert.insert(k,
                    'dif_z_score',
                    imp_fert['z_score_'+col_log[0]] - imp_fert['z_score_'+col_log[1]]
                   )
    
    imp_fert=z_score(imp_fert,'dif_z_score')
    
    borda=3
    
    imp_fert_corr=imp_fert[np.abs(imp_fert.z_score_dif_z_score)<3]
    imp_fert_err=imp_fert[np.abs(imp_fert.z_score_dif_z_score)>3]
    
    preço=imp_fert_corr.VL_FOB.sum()/imp_fert_corr.KG_LIQUIDO.sum()
    
    imp_fert_err=imp_fert_err.assign(preço=preço)
    
    imp_fert_err=imp_fert_err.assign(kg_err=1)
    
    imp_fert_err.loc[np.abs(imp_fert_err.z_score_log_VL_FOB)>=np.abs(imp_fert_err.z_score_log_KG_LIQUIDO),'kg_err']=0
    
    imp_fert_err_kg=imp_fert_err[imp_fert_err.kg_err==1]
    
    imp_fert_err_fob=imp_fert_err[imp_fert_err.kg_err==0]
    
    imp_fert_err_kg.KG_LIQUIDO=imp_fert_err_kg.VL_FOB/preço
    
    imp_fert_err_fob.VL_FOB=imp_fert_err_fob.KG_LIQUIDO*preço
    
    imp_fert=pd.concat([imp_fert_corr,imp_fert_err_kg,imp_fert_err_fob])
    
    imp_fert=imp_fert[col]
    
    return imp_fert

print('** Função corr_kg_fob_via_preço(imp_fert) disponível **')
```
