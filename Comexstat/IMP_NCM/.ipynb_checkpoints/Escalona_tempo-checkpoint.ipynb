```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

# Carrega a tabela imortação
i=imp_fert=pd.read_csv('/home/andre301267/git/BD/temp/imp_fert_ncm_quanti.csv').reset_index(drop=True)
i.shape
```

```python
mes_at=imp_fert[(imp_fert.CO_ANO==imp_fert.CO_ANO.max())].CO_MES.max()
bim=np.arange(1,7,.5).astype(int)
bim=pd.DataFrame({'CO_BIM':list(bim[int(mes_at/2):])+list(bim[:int(mes_at/2)])})
mes=pd.DataFrame({'CO_MES':np.arange(1,13)})
mes_bim=pd.concat([mes,bim],axis=1)

i=i.merge(mes_bim)
i[['CO_MES','CO_BIM']].drop_duplicates().sort_values('CO_MES')
```
