```python
import numpy as np
import pandas as pd

# Cria leitor do arquivo principal em pedaços
imp_fert=pd.read_csv('~/git/BD/temp/imp_fert_quanti.csv').assign(comex='imp')
imp_fert.columns
```

```python
# Cria leitor do arquivo principal em pedaços
exp_agr=pd.read_csv('~/git/BD/temp/exp_agr2.csv').assign(comex='exp')
ncm=pd.read_csv('~/git/BD/Comexstat/tab_complementares/NCM.csv',encoding='Latin-1',sep=';').assign(comex='exp')

exp_agr=exp_agr.merge(ncm)
exp_agr
```

```python
exp_agr=exp_agr.rename(columns={'NO_NCM_POR':'nick'})[imp_fert.columns]
```

```python
comex=pd.concat([imp_fert,exp_agr]).sort_values(by=['CO_ANO','CO_MES'])
```

```python
comex=comex[comex.SG_UF_NCM=='BA']
```

```python
comex.loc[comex.nick.str.lower().str.contains('soja'),'nick']='soja'
comex.loc[comex.nick.str.lower().str.contains('café'),'nick']='café'
comex.loc[comex.nick.str.lower().str.contains('milho'),'nick']='milho'

comex
```

```python
# Salva o resultado
comex.to_csv('~/git/BD/temp/comex.csv', index=False)
```
