```python
from sklearn.ensemble import RandomForestClassifier as RFC
RFC=RFC()
from sklearn.ensemble import RandomForestRegressor as RFR
RFR=RFR(max_depth=1)

import pandas as pd
import numpy as np

# Carrega a base de dados contendo os registros de comexportação de fertilizantes
comex=pd.read_csv('~/git/BD/temp/comex.csv')

# Modifica unidade de peso
comex.KG_LIQUIDO=comex.KG_LIQUIDO/10**6
comex=comex.rename(columns={'KG_LIQUIDO':'kton'})

comex.loc[comex.comex=='imp','nick']='ftlz'
comex=comex.groupby(by=['CO_ANO','CO_MES','nick'])[['kton','VL_FOB']].sum().reset_index()
comex_ftlz=comex[comex.nick=='ftlz']
comex_exp=comex[comex.nick!='ftlz']

q=pd.DataFrame()
a=1
for nick in comex_exp.nick.unique():
    t=comex[comex.nick==nick]
    t=t.rename(columns={'kton':'kton_ad_'+str(nick),'VL_FOB':'fob_ad_'+str(nick)})
    if a==1:
        q=t
        a+=1
    else:
        q=t.merge(q,how='outer')
q=q.fillna(0).drop(columns='nick')

ad=6
for ad1 in range(1,ad):
    q.CO_MES=q.CO_MES+1
    q.loc[q.CO_MES==13,'CO_ANO']=q[q.CO_MES==13]['CO_ANO']+1
    q.loc[q.CO_MES==13,'CO_MES']=1

q=q.groupby(by=['CO_ANO','CO_MES'])[q.columns[2:]].sum().reset_index()
i=comex_ftlz.merge(q,how='outer').fillna(0).drop(columns='nick')
i['date']=pd.to_datetime(i.CO_ANO.astype(str)+'/'+i.CO_MES.astype(str)+'/1').astype(int)
date_min=i.date.min()
date_max=i.date.max()
i[['CO_ANO','CO_MES']]=i[['CO_ANO','CO_MES']].astype('category')
i=i[['date','CO_MES','kton','kton_ad_soja','fob_ad_soja']]
i.head()
```

```python
# Formata mês como informação categórica
i.CO_MES=i.CO_MES.astype('category')
i=pd.get_dummies(i, prefix_sep='~')
i
```

```python
q=pd.DataFrame()
for a in range(120,len(i)):

    for train_size_mes in [120]:

        train=i.iloc[:a,:].tail(train_size_mes)
        
        if len(train)>=train_size_mes:

            test=i.iloc[a:a+1,:]
                        
            # Train
            X_train_tratamento=train.drop(columns=['kton'])
            X_train_testemunha=train.drop(columns=['kton','kton_ad_soja','fob_ad_soja'])
            y_train=train.kton
    
            # Test
            X_test_tratamento  =test.drop(columns=['kton'])
            X_test_testemunha=test.drop(columns=['kton','kton_ad_soja','fob_ad_soja'])
            y_test=test.kton
            RFR.fit(X_train_tratamento,y_train)
            
            test_pred=test     .assign(kton_pred_tratamento  =RFR.predict(X_test_tratamento  ).astype(int))
            
            RFR.fit(X_train_testemunha,y_train)
            
            test_pred=test_pred.assign(kton_pred_testemunha=RFR.predict(X_test_testemunha).astype(int))

            test_pred=test_pred.assign(train_size_anos=train_size_mes/12)
            
            q=pd.concat([q,test_pred])

q['erro_tratamento']=np.abs(q.kton-q.kton_pred_tratamento)
q['erro_testemunha']=np.abs(q.kton-q.kton_pred_testemunha)
```

```python
q.to_csv('~/git/BD/temp/q.csv', index=False)
```
