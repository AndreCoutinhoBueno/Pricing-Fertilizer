```python
from sklearn.ensemble import RandomForestClassifier as RFC
RFC=RFC()
from sklearn.ensemble import RandomForestRegressor as RFR
RFR=RFR(max_depth=None)

import pandas as pd
import numpy as np

# Carrega a base de dados contendo os registros de comexportação de fertilizantes
comex=pd.read_csv('~/git/BD/temp/comex.csv')
comex
```

```python
comex=comex[comex.nick.isin(['MOP_k','soja'])]

comex=comex[comex.comex=='imp'].merge(comex[comex.comex=='exp'],on=['CO_ANO','CO_MES'])\
.rename(columns={'KG_LIQUIDO_x':'kg_ftlz','KG_LIQUIDO_y':'kg_exp','VL_FOB_x':'fob_ftlz','VL_FOB_y':'fob_exp'})

comex=comex[comex.columns[[0,1,5,6,10,11]]]
```

```python
comex=comex.assign(rel=(comex.fob_ftlz/comex.kg_ftlz)/(comex.fob_exp/comex.kg_exp))
comex
```

```python
c1=comex.loc[:,:'kg_ftlz']
c2=comex[comex.columns[[0,1,3,4,5,6]]]

ad=1
for ad_ in range(1,ad+1):
        
    c2.CO_MES=c2.CO_MES+1
    c2.loc[c2.CO_MES==13,'CO_ANO']=c2.loc[c2.CO_MES==13]['CO_ANO']+1
    c2.loc[c2.CO_MES==13,'CO_MES']=1


c2=c2.rename(columns={'fob_ftlz':'fob_ftlz_ad','kg_exp':'kg_exp_ad','fob_exp':'fob_exp_ad','rel':'rel_ad'})
comex=c1.merge(c2,on=['CO_ANO','CO_MES'])

i=comex

i
```

```python
i['date']=pd.to_datetime(i.CO_ANO.astype(str)+'/'+i.CO_MES.astype(str)+'/1')
date_min=i.date.min()
date_max=i.date.max()
```

```python
# Formata mês como informação categórica
i.CO_ANO=i.CO_ANO.astype('category')
i.CO_MES=i.CO_MES.astype('category')
```

```python
i.columns
```

```python
i=i.groupby(by=['CO_ANO','CO_MES'],observed=False)[i.columns[2:-1]].sum().reset_index()
i['date']=pd.to_datetime(i.CO_ANO.astype(str)+'/'+i.CO_MES.astype(str)+'/1')
i=i[(i.date>=date_min)&(i.date<=date_max)]
i.date=i.date.astype(int)
```

```python
i=pd.get_dummies(i, prefix_sep='~')
i
```

```python
q=pd.DataFrame()
for a in range(120,len(i)):

    for train_size_mes in [120]:

        train=i.iloc[:a,:].tail(train_size_mes)
        
        if len(train)>=train_size_mes:
            train=train[train.kg_ftlz>0]

            test=i.iloc[a:a+1,:]

            if (len(train)>3)&(len(test)>0):
                
                # Train
                X_train_tratamento=train.drop(columns=['kg_ftlz','fob_ftlz_ad'])
                X_train_testemunha=train.drop(columns=['kg_ftlz','fob_ftlz_ad','fob_exp_ad','kg_exp_ad','rel_ad'])
                y_train=train.kg_ftlz
        
                # Test
                X_test_tratamento  =test.drop(columns=['kg_ftlz','fob_ftlz_ad'])
                X_test_testemunha=test.drop(columns=['kg_ftlz','fob_ftlz_ad','fob_exp_ad','kg_exp_ad','rel_ad'])
                y_test=test.kg_ftlz
                RFR.fit(X_train_tratamento,y_train)
                
                test_pred=test     .assign(kg_ftlz_pred_tratamento  =RFR.predict(X_test_tratamento  ).astype(int))
                
                RFR.fit(X_train_testemunha,y_train)
                
                test_pred=test_pred.assign(kg_ftlz_pred_testemunha=RFR.predict(X_test_testemunha).astype(int))
    
                test_pred=test_pred.assign(train_size_anos=train_size_mes/12)
                
                q=pd.concat([q,test_pred])

q['erro_tratamento']=np.abs(q.kg_ftlz-q.kg_ftlz_pred_tratamento)
q['erro_testemunha']=np.abs(q.kg_ftlz-q.kg_ftlz_pred_testemunha)
```

```python
q.to_csv('~/git/BD/temp/q.csv', index=False)
```
