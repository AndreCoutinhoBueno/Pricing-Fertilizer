```python
import numpy as np
import pandas as pd
```

```python
imp_fert=pd.read_csv('~/git/BD/temp/imp_fert_quanti.csv').sort_values(by=['CO_ANO','CO_MES']).reset_index(drop=True)
```

```python
filt_ncm=imp_fert.groupby('CO_NCM').KG_LIQUIDO.sum().sort_values(ascending=False).head(2).reset_index().CO_NCM.values[:1]
imp_fert=imp_fert[imp_fert.CO_NCM.isin(filt_ncm)]

filt_uf=imp_fert.groupby('SG_UF_NCM').KG_LIQUIDO.sum().sort_values(ascending=False).head(2).reset_index().SG_UF_NCM.values[:1]
imp_fert=imp_fert[imp_fert.SG_UF_NCM.isin(filt_uf)]
```

```python
imp_fert=imp_fert.loc[:,:'VL_FRETE']
imp_fert[['CO_ANO','CO_MES','CO_NCM','SG_UF_NCM']]=imp_fert[['CO_ANO','CO_MES','CO_NCM','SG_UF_NCM']].astype('category')
imp_fert=imp_fert.groupby(observed=False,by=['CO_ANO','CO_MES','CO_NCM','SG_UF_NCM'])[['KG_LIQUIDO','VL_FOB','VL_FRETE']].sum().reset_index()
imp_fert[['CO_ANO','CO_MES']]=imp_fert[['CO_ANO','CO_MES']].astype(int)
imp_fert[['CO_NCM','SG_UF_NCM']]=imp_fert[['CO_NCM','SG_UF_NCM']].astype(str)
```

```python
imp_fert.insert(0,'date',pd.to_datetime(imp_fert.CO_ANO.astype(str)+'/'+imp_fert.CO_MES.astype(str)+'/1'))

imp_fert=imp_fert.set_index('date')

imp_fert2=imp_fert.shift(1,freq='ME').shift(1,freq='D').reset_index().rename(columns={'KG_LIQUIDO':'kg_ad',\
                                                                                     'VL_FOB':'fob_ad',\
                                                                                     'VL_FRETE':'frete_ad'})

imp_fert2.CO_ANO=imp_fert2.date.dt.year
imp_fert2.CO_MES=imp_fert2.date.dt.month

imp_fert=imp_fert.reset_index()

imp_fert2
```

```python
imp_fert3=imp_fert.merge(imp_fert2)
imp_fert3.insert(0,'ncm_uf',imp_fert3.CO_NCM+'_'+imp_fert3.SG_UF_NCM+'_kg_ad')

imp_fert3=imp_fert3[['date','ncm_uf','kg_ad','fob_ad','frete_ad']]
imp_fert3
```

```python
q=pd.DataFrame()
a=0
for date in imp_fert3.date.unique():
        
    imp_fert3_t=imp_fert3[imp_fert3.date==date].drop(columns='date').set_index('ncm_uf').transpose().iloc[0:1]
    
    imp_fert3_t.insert(0,'date',date)

    
    
    q=pd.concat([q,imp_fert3_t])

imp_fert4=q
```

```python
imp_fert5=imp_fert.merge(imp_fert2,how='outer').merge(imp_fert4,how='outer').fillna(0)

# Elimina colunas target potencial, que em estágios futuros, pode deixar de ser feito
imp_fert5=imp_fert5.drop(columns=['VL_FOB','VL_FRETE','kg_ad'])
imp_fert5.date=imp_fert5.date.astype(int)
imp_fert5.head(1)
```
