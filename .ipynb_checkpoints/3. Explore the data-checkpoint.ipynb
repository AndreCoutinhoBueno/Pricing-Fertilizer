# Fertilizantes  

O foco do projeto são as importações relativas as Unidades da Receita Federal dos portos de Salvador e Aratu na Bahia, destinadas a esta mesma Unidade Federativa.

```python
urf=pd.read_csv('~/git/BD/Comexstat/URF.csv',encoding='Latin-1',sep=';')
urf_salvador_aratu=urf[(urf.NO_URF.str.contains('ARATU'))|(urf.NO_URF.str.contains('SALVADOR'))].CO_URF
urf_salvador_aratu
```

```python
i=pd.read_csv('~/git/Pricing-Fertilizer-take-Rains-into-account/temp/imp1.csv')
i=i[(i.CO_URF.isin(urf_salvador_aratu))&(i.SG_UF_NCM=='BA')]
i.head()
```

Como a coluna SG_UF_NCM ficou apenas com uma única categoria, 'BA', e como não há perspectiva de aproveitar a informação sobre unidade de  Receita Federal, essas duas colunas serão descartadas.

```python
i.groupby('CO_VIA')['KG_LIQUIDO'].sum().sort_values()
```

```python
i.groupby('CO_UNID')['KG_LIQUIDO'].sum().sort_values()
```

Como as colunas 'CO_VIA' e 'CO_UNID' apresentam praticamente todos os valores em uma única categoria, elas serão também  descartadas.

A coluna 'QT_STAT' representa a quantidade importada medidas em outras unidades de que não sejam o 'KG_LIQUIDO'. Isso não é de interesse do projeto e a coluna também será descartada.

```python
i.drop(columns=['SG_UF_NCM','CO_URF','CO_UNID','CO_VIA','QT_ESTAT'], inplace=True)
```

O arquivo possui dois registros sem peso maior que zero. Estes mesmos registros apresentam valor FOB irrisório e por isso serão descatados.

```python
i=i[i.KG_LIQUIDO>0]
```

Numa primeira tentativa de atingir o objetivo, não levaremos em análise os diferentes NCMs e países de origem do registro de importação. Assim, a coluna 'CO_NCM' e 'CO_PAIS' serão descrtadas e os valores agrupados. Os valores monetários também serao descartados.

```python
i=i.groupby(by=['CO_ANO','CO_MES'])['KG_LIQUIDO'].sum().reset_index()
```

```python
sns.displot(data=i,x='KG_LIQUIDO',log_scale=True).fig.suptitle('Importação de Fertilizantes na Bahia',y=1.01)
```

Os valores menores do que 10**6.3 serão descartados.

```python
i=i[i.KG_LIQUIDO>10**6.3]
```

Um registro apresenta peso muito acima da normalidade.

```python
i[i.KG_LIQUIDO==i.KG_LIQUIDO.max()]
```

Ele será substituido pela interpolação dos valores do mês no ano anterior e posterior ao ocorrido.

```python
ano_erro=i[i.KG_LIQUIDO==i.KG_LIQUIDO.max()].CO_ANO.values[0]

ano_corr_anterior=i[i.CO_ANO==ano_erro+1].CO_ANO.values[0]

ano_corr_posterior=i[i.CO_ANO==ano_erro-1].CO_ANO.values[0]

mes_erro=i[i.KG_LIQUIDO==i.KG_LIQUIDO.max()].CO_MES.values[0]

val_corr=i[i.CO_ANO.isin([ano_corr_anterior,ano_corr_posterior])].KG_LIQUIDO.mean()

i.loc[i.KG_LIQUIDO==i.KG_LIQUIDO.max(),'KG_LIQUIDO']=val_corr
```

```python
sns.displot(data=i,x='KG_LIQUIDO',log_scale=True, bins=6).fig.suptitle('Importação de Fertilizantes na Bahia',y=1.01)
```

```python
sns.displot(data=i,x='KG_LIQUIDO',log_scale=True,col='CO_MES',col_wrap=4, bins=6).fig.suptitle('Importação de Fertilizantes na Bahia',y=1.01)
```

```python
i['date']=pd.to_datetime(i.CO_ANO.astype(str)+'/'+i.CO_MES.astype(str)+'/1')
```

```python
sns.relplot(data=i,x='date',y='KG_LIQUIDO',col='CO_MES',col_wrap=4).fig.suptitle('Importação de Fertilizantes na Bahia',y=1.01)
```
