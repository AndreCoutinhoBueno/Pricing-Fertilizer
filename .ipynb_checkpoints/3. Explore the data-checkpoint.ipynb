```python
g=sns.displot(height=2, aspect=1.618,data=i,x='Kg',log_scale=True)
g.fig.suptitle('Importação de Fertilizantes na Bahia',y=1.1)
plt.show()
```

Um registro apresenta peso muito acima da normalidade, próximo de 10**9 kg.

```python
i[i.Kg==i.Kg.max()]
```

Ele será substituido por interpolação dos valores do mesmo mês porém do ano anterior e do ano posterior.

```python
ano_erro=i[i.Kg==i.Kg.max()].Ano.values[0]
ano_corr_anterior=i[i.Ano==ano_erro+1].Ano.values[0]
ano_corr_posterior=i[i.Ano==ano_erro-1].Ano.values[0]
mes_erro=i[i.Kg==i.Kg.max()].Mês.values[0]
val_corr=i[i.Ano.isin([ano_corr_anterior,ano_corr_posterior])].Kg.mean()
i.loc[i.Kg==i.Kg.max(),'Kg']=val_corr
```

A distribição dos pesos registrados ficou como a seguir e foi considerada normal.

```python
sns.displot(height=2, aspect=1.618,data=i,x='Kg',log_scale=True, bins=6).fig.suptitle('Importação de Fertilizantes na Bahia',y=1.1)
plt.show()
```

```python
# Cria luna com a data completa e no formato apropriado
i['date']=pd.to_datetime(i.Ano.astype(str)+'/'+i.Mês.astype(str)+'/1')
```

Para conferência ampla, a seguir podemos ver as ocorrências por mês ao longo dos anos.

```python
g=sns.relplot(height=2, aspect=1.618,data=i,x='date',y='Kg',col='Mês',col_wrap=6)
g.fig.suptitle('Importação de Fertilizantes na Bahia',y=1.1)
#g.set_xticklabels(rotation=90)
plt.tight_layout()
```

A tabela e passada ao formato de martiz esparsa, o que permitirá ela ser fundida adiante com a tabela de chuvas, que passará pelo mesmo processo, evitando a perda de registros.

```python
# Martiz Esparsa
i_ini=i.date.min()
i_fim=i.date.max()
i[['Ano','Mês']]=i[['Ano','Mês']].astype('category')
i2=i.groupby(by=['Ano','Mês']).Kg.sum().reset_index()
i2['date']=pd.to_datetime(i2.Ano.astype(str)+'/'+i2.Mês.astype(str)+'/1')
i2=i2[(i2.date>=i_ini)&(i2.date<=i_fim)]
```

Nova coluna conténdo a regressão linear polinomial de terceiro graus da coluna peso. O objetivo é a eliminação de tendência a longo prazo destes valores de forma a permitir comparação com os índices de chuva, que passará pelo mesmo processo.

```python
i2['KG_fit']=np.poly1d(np.polyfit(i2.date.astype(int),i2.Kg, deg=4))(i2.date.astype(int))
```

Nova coluna é criada conténdo o valor do peso das importações descontado o valor da tendência no momento.

```python
i2['KG_resi']=i2.Kg-i2.KG_fit
```

Para permitir a apresentação gráfica adequada, a tabela com importações é duplicada e passada para o formato longo.

```python
i3=pd.concat([
    i2[['date','Kg']].assign(tipo='real'   ).rename(columns={'Kg':'KG'}),
    i2[['date','KG_fit'    ]].assign(tipo='tendência').rename(columns={'KG_fit'    :'KG'})
    ])
```

Apresenta graficamente a tendência das importações, que será eliminada como descrito anteriormente.

```python
sns.relplot(height=2, aspect=1.618,data=i3,x='date',y='KG',hue='tipo').fig.suptitle('Importação de Fertilizantes na Bahia',y=1.1)
plt.xticks(rotation=90)
plt.show()
```

### 2.Chuvas

```python
# Importa as bibliotecas necessárias
%run '~/git/Bibliotecas.ipynb'
```

```python
# Carrega a tabela com histórico das chuvas no oeste baiano
c=pd.read_csv('/home/andre301267/git/Pricing-Fertilizer-take-Rains-into-account/temp/chuvas1.csv')
```

Seleciona colunas de interesse, simplifica os nomes das colunas, unifica colunas de data que estavam em duplicidade.

```python
c=c[c.columns[[20,0,21,1,22,2]]]
c.columns=['Estação','date1','date2','hora1','hora2','mm']
c.date1=pd.to_datetime(c.date1+'-'+c.hora1)
c.date2=pd.to_datetime(c.date2+'-'+c.hora2)
c['date']=pd.to_datetime(pd.concat([c.date1.dropna(),c.date2.dropna()]), utc=True)
c['Ano']=c.date.dt.year
c['Mês']=c.date.dt.month
c['Dia']=c.date.dt.day
c['Hora']=c.date.dt.hour
c=c[c.columns[[0,6,7,8,9,10,5]]]
c.head(1)
```

Substitui valores inválidos e formatação numérica para o formato adequado.

```python
c.loc[c.mm=='-9999','mm']=np.nan
c.mm=c.mm.str.replace(',','.')
c.mm=c.mm.astype('float')
```

Não há valores duplicados.

```python
c.duplicated().sum()
```

Há aproximadamente cem mil falhas nos registros de chuvas por hora.

```python
c.isna().sum()
```

Os registros com falha e sem falhas são separados em tabelas próprias.

```python
c_na=c[(c.mm.isna())]
c_not_na=c[~(c.mm.isna())]
```

```python
print(f"As importações com falha no registro de chuvas correspondem a {round(100*len(c_na)/len(c))} % do total")
```

A seguir se modifica a tabela de falhas de registro de forma a permitir a análise gráfica da distribuição das falhas ao longo das categorias existentes, com objetivo de verificar se há algum sentido lógico para as ocorrências.

```python
w=pd.concat([\
c_na.groupby('Ano' ).Mês.count().reset_index().rename(columns={'Mês':'count','Ano':' '}).assign(tipo='Ano' ),
c_na.groupby('Mês' ).Ano.count().reset_index().rename(columns={'Ano':'count','Mês':' '}).assign(tipo='Mês' ),
c_na.groupby('Dia'    ).Ano.count().reset_index().rename(columns={'Ano':'count','Dia':' '}).assign(tipo='Dia'    ),
c_na.groupby('Hora'   ).Ano.count().reset_index().rename(columns={'Ano':'count','Hora':' '}).assign(tipo='Hora'   ),
c_na.groupby('Estação').Ano.count().reset_index().rename(columns={'Ano':'count','Estação':' '}).assign(tipo='Estação')])

for t in w.tipo.unique():
    w_t=w[w.tipo==t]
    sns.catplot(height=1,aspect=4*1.6,data=w_t,x=' ',y='count',kind='bar').fig.suptitle(t,y=1.1)
    plt.xticks(rotation=90)
    if t =='Estação':
        plt.xticks(rotation=22.5)
    elif t=='Mês':
        plt.xticks(rotation=0)
print(f"\n\n{round(100*len(c_not_na[c_not_na.mm!=0])/len(c_not_na))} % das horas registradas apresentam falhas na marcação do volume de chuva em 'mm'.")
print('\n\n                  Horas com Falha no Registro de Chuvas\n\n\n')
plt.show()

print('\nConclsão: As falhas no registro de chuva aparentam estar normalmente distribuidas em todas as categorias,\
\nnão sendo identificado nenhum efeito causal entre elas.\n\n\n\n')
```

A seguir analizaremos a normalidade da distribuição dos registros de chuva. A cada aprovação de normalidade, se elimina a menor subdivisão de tempo, reagrupando, até chegar na subdvisão de tempo por mês, que é a medida de tempo foco deste projeto.

```python
# Cria nova coluna com volume de chuva em unidade logarítmica
c_not_na['mm_hora_log']=np.log10(c_not_na.mm)
```

```python
# Plota volumes logarítmos de chuva por hora em gráfico com escala logarítmica (duplo logaritmo)
sns.displot(height=2, aspect=1.618,data=c_not_na[c_not_na.mm_hora_log!=0],x='mm_hora_log',log_scale=True, bins=4).fig.suptitle('Chuvas no Oeste Baiano',y=1.1)
plt.show()
```

A distribuição de dos volumes de chuva por hora foram considerados normais.
A seguir se arupa os valores por dia.

```python
c_dia=c_not_na.groupby(by=c_not_na.columns[[0,1,2,3,4]].tolist())[['mm']].sum().reset_index()
c_dia.head(1)
```

Recria-se a coluna com volume de chuva logarítmico.

```python
c_dia['mm_dia_log']=np.log10(c_dia.mm)
```

```python
# Plota-se novamente os volumes logarítmicos de chuva em gráfico com escala logarítmica
sns.displot(height=2, aspect=1.618,data=c_dia[c_dia.mm_dia_log!=0],x='mm_dia_log',log_scale=True, bins=4).fig.suptitle('Chuvas no Oeste Baiano',y=1.1)
plt.show()
```

Nos dois gráficos apresentados acima podemos notar:  

> Devido a grande amplitude nos volumes de chuvas registrados, para possibilitar a visualização da curva normal, foi necessário aplicar duas vezes seguidas a transformação logarítmica aos volumes de chuva:
> * primeiro: transformando diretamente o valor medido para logarítmo
> * segundo: utilizando no gráfico a escala logarítmica

Agrupa-se a tabela por mês

```python
c_mes=c_dia.groupby(by=['Estação','Ano','Mês']).mm.sum().reset_index().rename(columns={'mm':'mm_mes'})
```

Com os agrupamentos até aqui procedidos, não é mais necessário a dupla transofrmação logaritmica, mas ainda é necessário a escala logarítmica no gráfico, para se visualizar a curva normal d distribuição de chuvas por mês.

```python
sns.displot(height=2, aspect=1.618,data=c_mes[c_mes.mm_mes!=0],x='mm_mes',log_scale=True, bins=4).fig.suptitle('Chuvas no Oeste Baiano',y=1.1)
plt.show()
```

Para simplificar, levaremos em conta apenas um linha do tempo de registros de chuva, sendo a seguir eleiminada a subdivisão referente as estações meteorológicas e agrupando os dados utilizando os valores medianos entre as estações.

```python
c=c_mes.groupby(by=['Ano','Mês']).mm_mes.median().reset_index()
```

```python
sns.displot(height=2, aspect=1.618,data=c[c.mm_mes!=0],x='mm_mes',log_scale=True,bins=4).fig.suptitle('Falhas no Registro de Chuvas',y=1.1)
plt.show()
```

Após realizar todos os agrupamentos necessários até alcançar a unidade de tempo desejada, por fim, os dados são considerados normais.

```python
# Cria coluna com data completa e no formato adequado
c['date']=pd.to_datetime(c.Ano.astype(str)+'/'+c.Mês.astype(str)+'/1')
```

```python
# Cria matriz esparsa, com o mesmo objetivo do que foi feito com a tabela de imortações
c_ini=c.date.iloc[0]
c_fim=c.date.iloc[-1]
c[['Ano','Mês']]=c[['Ano','Mês']].astype('category')
c=c.groupby(by=['Ano','Mês']).mm_mes.sum().reset_index()
c[['Ano','Mês']]=c[['Ano','Mês']].astype('int')
c['date']=pd.to_datetime(c.Ano.astype(str)+'/'+c.Mês.astype(str)+'/1')
c=c[(c.date>=c_ini)&(c.date<=c_fim)]
c=c.rename(columns={'mm_mes':'mm'})
c=c.sort_values(by=['date'])
c
```

### 3. Funde as tabelas de importações e de chuvas

```python
i_c=i.merge(c)
```

Para comparação das quantidades imortadas com os volumes de chuva, duas novas colunas são criadas com estes valores numa mesma escala de 0 a 1:

```python
from sklearn.preprocessing import MinMaxScaler as mms
mms=mms()
i_c_mms=pd.DataFrame(mms.fit_transform(i_c[['KG_resi','mm']]))#,'mm_log']]))
i_c_mms.columns=['kg_resi_mms','mm_mms']#,'mm_log_mms']
i_c=pd.concat([i_c,i_c_mms],axis=1)
```

Para seguir facilitando a comparação entre peso importado e volume de chuvas, criamos a seguir novas colunas conténdo as médias dessas grandezas a cada mês:

```python
# Cria médias mensais
kg_resi_m = i_c.groupby('Mês').kg_resi_mms.mean().reset_index().rename(columns={'kg_resi_mms':'kg_resi_mms_mean'})
mm_m      = i_c.groupby('Mês').mm_mms     .mean().reset_index().rename(columns={'mm_mms'     :'mm_mms_mean'     })
```

Funde-se as tabelas de importações e de registros dos voumes de chuvas

```python
i_c=i_c.merge(kg_resi_m).merge(mm_m)
```

```python
# Cria-se novas colunas conténdo os desvios das médias de importações e de chuvas
i_c['kg_resi_mms_dif'] = i_c.kg_resi_mms - i_c.kg_resi_mms_mean
i_c['mm_mms_dif']      = i_c.mm_mms      - i_c.mm_mms_mean
```

Com certeza as chuvas nas regiões agrícolas não tem como modificar imediatamente o desembarque de fertilizantes no porto. Da mesma forma, não tem como impactar de produtos que já estejam a caminho pelo mar. O importador teria dificuldade também de, devido as chuvas, modificar o volume já encomendado a um fabricante.  

Pata tentar identificar o intervalo de tempo de  uma ocorrência de uma chuva irregular sobre o peso de uma importação, a seguir apresentamos análise de impacto desse fato para uma faixa de tempo de separação entre estes dois eventos:

```python
from sklearn import metrics

for ad in range(0,13):
    kg=i_c[['kg_resi_mms_dif']]
    mm=i_c[['mm_mms_dif']]
    mm.index=mm.index+ad
    i_c_1=pd.concat([kg,mm],axis=1).dropna()
    print(f"Adiantando {ad} meses-Média de Erro Absoluto={round(100*metrics.mean_absolute_error(i_c_1['mm_mms_dif'],i_c_1['kg_resi_mms_dif']),1)}")

print()
```

Transforma-se a tabela para o formato longo, de forma a ser utilizado na construção dos gráficos a serem apresentados em seguida:

```python
# Produz i_c_l
a =i_c[['Ano','Mês','date','kg_resi_mms_dif']].assign(tipo='kg_resi_mms_dif').rename(columns={'kg_resi_mms_dif':'mms_dif'})
b =i_c[['Ano','Mês','date','mm_mms_dif'     ]].assign(tipo='mm_mms_dif'     ).rename(columns={'mm_mms_dif'     :'mms_dif'})
i_c_l=pd.concat([a,b])
```

```python
# Apresenta os gráfico de comparação das médias escalonadas entre as importações e as chuvas a cada ano ao longo dos meses
sns.relplot(height=3, aspect=1/1.618,kind='line',data=i_c_l,x='Mês',y='mms_dif',hue='tipo', col='Ano',col_wrap=6).fig.suptitle("Proporção entre Chuvas e Importação de Fertilizantes",y=1.1)
plt.show()
```

```python
# Apresenta os gráfico de comparação das médias escalonadas entre as importações e as chuvas a cada mês ao longo dos anos
sns.relplot(height=3, aspect=1/1.618,kind='line',data=i_c_l,x='Ano',y='mms_dif',hue='tipo', col='Mês',col_wrap=6).fig.suptitle("Proporção entre Chuvas e Importação de Fertilizantes",y=1.1)
plt.show()
```
